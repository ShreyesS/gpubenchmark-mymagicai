```python
import time
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
```

```python
device = torch.device("cuda")
```

```python
# Load LLM model and tokenizer
#model_name = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"
#llm = AutoModelForCausalLM.from_pretrained("TheBloke/Llama-2-7B-GGML", model_file = "llama-2-7b.ggmlv3.q2_K.bin")

# models = TinyLlama/TinyLlama-1.1B-Chat-v1.0, TheBloke/Mistral-7B-v0.1-AWQ, bigscience/bloom-560m

#TheBloke/Mistral-7B-v0.1-AWQ, 

model = AutoModelForCausalLM.from_pretrained("TinyLlama/TinyLlama-1.1B-Chat-v1.0")#.to("cuda")
tokenizer = AutoTokenizer.from_pretrained("TinyLlama/TinyLlama-1.1B-Chat-v1.0")
```

```python
input_prompt = "Once upon a time, "

batch_sizes = [x for x in range(1,100)]
```

```python
print(next(model.parameters()).device)
```

```python
for batch_size in batch_sizes:
    # Generate input sequences
    inputs = tokenizer.encode(input_prompt, return_tensors="pt").repeat(batch_size, 1)#.to(device)
    
    #inputs = inputs.to("cuda")
    # Warm-up inference
    #_ = llama_llm.generate(inputs)

    # Measure inference time
    start_time = time.time()
    outputs = model.generate(inputs)
    end_time = time.time()
    runtime = end_time - start_time

    # Calculate latency and throughput
    latency = runtime / batch_size
    throughput = batch_size / runtime

    print(f"Batch Size: {batch_size}")
    print(f"Runtime: {runtime:.2f} seconds")
    print(f"Latency: {latency:.4f} seconds per sample")
    print(f"Throughput: {throughput:.2f} samples per second")
    print("=" * 50)
```
